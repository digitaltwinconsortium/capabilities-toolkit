## Step 5: Core Visualization Function

The main visualization function creates professional CPT diagrams. Key features:

### Visual Elements:
- **Grid Layout**: 6Ã—8 grid with proper spacing
- **Color Coding**: Category-specific colors for recognition
- **Highlighting**: Required capabilities at full opacity, others faded
- **Typography**: Professional fonts with proper sizing
- **Legend**: Category reference for easy interpretation

### Customization Options:
- **Figure Size**: Adjustable for different output needs
- **Save Options**: High-resolution PNG export
- **Title/Description**: Custom text for specific use cases
- **Legend Control**: Show/hide category legend

This function is the core of our CPT generator and will be used for all visualizations.

````python
# Core CPT Visualization Function

def create_cpt_visualization(required_capabilities=None, title=None, description=None, 
                           figsize=(16, 12), save_path=None, show_legend=True, 
                           config_name=None):
    """
    Create a professional CPT visualization
    
    Parameters:
    - required_capabilities: list of capability IDs to highlight
    - title: custom title for the visualization
    - description: custom description text
    - figsize: tuple for figure dimensions
    - save_path: path to save PNG file
    - show_legend: whether to show category legend
    - config_name: use predefined configuration
    """
    
    # Use predefined config if specified
    if config_name and config_name in configurations:
        config = configurations[config_name]
        required = config['required']
        plot_title = config['title']
        plot_desc = config['description']
    else:
        required = required_capabilities or []
        plot_title = title or 'AI Agent Capabilities Periodic Table'
        plot_desc = description or 'Custom Configuration'
    
    # Create the figure
    fig, ax = plt.subplots(figsize=figsize)
    ax.set_xlim(0, 7)
    ax.set_ylim(0, 9)
    ax.set_aspect('equal')
    
    # Remove axes for clean appearance
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    
    # Add title and description
    fig.suptitle(plot_title, fontsize=20, fontweight='bold', y=0.95)
    ax.text(3.5, 8.5, plot_desc, ha='center', va='center', fontsize=14, 
            style='italic', color='#666666')
    
    # Draw all capabilities
    for cap in capabilities:
        # Determine if this capability is required for this use case
        is_required = cap['id'] in required
        
        # Get colors for this category
        colors = category_colors[cap['category']]
        
        # Set opacity based on requirement (required=100%, others=30%)
        alpha = 1.0 if is_required else 0.3
        
        # Calculate position (flip Y axis for proper display)
        x = cap['col'] - 0.4
        y = 9 - cap['row'] - 0.4
        
        # Create capability rectangle
        rect = Rectangle((x, y), 0.8, 0.8, 
                        facecolor=colors['fill'], 
                        edgecolor=colors['border'],
                        linewidth=2 if is_required else 1,
                        alpha=alpha)
        ax.add_patch(rect)
        
        # Add capability text
        # Line 1: Capability ID (bold)
        ax.text(x + 0.4, y + 0.6, cap['id'], 
               ha='center', va='center', 
               fontsize=9, fontweight='bold',
               color=colors['text'], alpha=alpha)
        
        # Line 2: Capability name
        ax.text(x + 0.4, y + 0.25, cap['name'], 
               ha='center', va='center', 
               fontsize=7, 
               color=colors['text'], alpha=alpha)
    
    # Add category legend if requested
    if show_legend:
        legend_elements = []
        category_names = {
            'PK': 'Perception & Knowledge',
            'CG': 'Cognition & Reasoning',
            'LA': 'Learning & Adaptation',
            'AE': 'Action & Execution',
            'IC': 'Interaction & Collaboration',
            'GS': 'Governance & Safety'
        }
        
        for cat, name in category_names.items():
            colors = category_colors[cat]
            legend_elements.append(
                patches.Patch(facecolor=colors['fill'], 
                            edgecolor=colors['border'],
                            label=f"{cat}: {name}")
            )
        
        ax.legend(handles=legend_elements, loc='lower center', 
                 bbox_to_anchor=(0.5, -0.15), ncol=3, fontsize=10)
    
    plt.tight_layout()
    
    # Save if path provided
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', 
                   facecolor='white', edgecolor='none')
        print(f"ðŸ’¾ CPT saved to: {save_path}")
    
    plt.show()
    
    # Print summary statistics
    total_caps = len(capabilities)
    required_count = len(required)
    coverage_percent = (required_count / total_caps) * 100
    
    print(f"\nðŸ“Š CPT Summary:")
    print(f"Configuration: {plot_title}")
    print(f"Required Capabilities: {required_count}/{total_caps} ({coverage_percent:.1f}%)")
    
    # Count categories represented
    represented_categories = set()
    for cap_id in required:
        for cap in capabilities:
            if cap['id'] == cap_id:
                represented_categories.add(cap['category'])
                break
    
    print(f"Categories represented: {len(represented_categories)}/6")
    
    return fig, ax

print("âœ… Core visualization function defined!")
print("ðŸŽ¨ Ready to create professional CPT diagrams")
````

## Step 8: Maturity Level Comparison

Agent maturity levels (L0-L4) represent increasing sophistication in AI agent capabilities:

- **L0 - Static Automation**: Pre-programmed responses, no learning
- **L1 - Conversational Agents**: Natural language interaction, basic context
- **L2 - Procedural Workflow**: Multi-step tasks, tool integration
- **L3 - Cognitive Autonomous**: Self-directed planning, learning from experience
- **L4 - Multi-Agent Systems**: Collaborative intelligence, distributed consensus

This visualization shows how capability requirements evolve across maturity levels.

````python
# Maturity Level Comparison Visualization

def create_maturity_comparison(figsize=(20, 12), save_path=None):
    """
    Create a side-by-side comparison of agent maturity levels L0-L4
    """
    
    # Define capability sets for each maturity level
    maturity_levels = {
        'L0 - Static Automation': {
            'capabilities': ['PK.KB', 'AE.TX', 'GS.SE'],
            'description': 'Pre-programmed responses\nRule-based systems'
        },
        'L1 - Conversational': {
            'capabilities': ['IC.NL', 'IC.DM', 'PK.KB', 'PK.CX', 'AE.CX', 'GS.SE', 'GS.PR'],
            'description': 'Natural language interaction\nBasic context management'
        },
        'L2 - Procedural Workflow': {
            'capabilities': ['CG.PL', 'AE.TX', 'AE.TL', 'IC.AC', 'IC.CL', 'IC.RB', 'GS.MO'],
            'description': 'Multi-step task execution\nTool integration'
        },
        'L3 - Cognitive Autonomous': {
            'capabilities': ['CG.RS', 'CG.PS', 'CG.DC', 'LA.MM', 'LA.AD', 'PK.OB', 'AE.CG', 'GS.EX'],
            'description': 'Self-directed planning\nLearning from experience'
        },
        'L4 - Multi-Agent Systems': {
            'capabilities': ['IC.CS', 'IC.CF', 'IC.DS', 'IC.SI', 'IC.ES', 'IC.MB', 'GS.TC', 'GS.RL'],
            'description': 'Collaborative intelligence\nDistributed consensus'
        }
    }
    
    # Create subplot layout
    fig, axes = plt.subplots(1, 5, figsize=figsize)
    fig.suptitle('AI Agent Maturity Levels Comparison (L0-L4)', 
                 fontsize=24, fontweight='bold', y=0.95)
    
    # Draw each maturity level
    for idx, (level, config) in enumerate(maturity_levels.items()):
        ax = axes[idx]
        _draw_mini_cpt(ax, config['capabilities'], level, config['description'])
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight', 
                   facecolor='white', edgecolor='none')
        print(f"ðŸ’¾ Maturity comparison saved to: {save_path}")
    
    plt.show()
    
    # Print capability evolution summary
    print("\nðŸ“Š Capability Evolution Across Maturity Levels:")
    for level, config in maturity_levels.items():
        cap_count = len(config['capabilities'])
        print(f"{level}: {cap_count} capabilities")
        
        # Show category distribution
        cat_counts = {}
        for cap_id in config['capabilities']:
            for cap in capabilities:
                if cap['id'] == cap_id:
                    cat = cap['category']
                    cat_counts[cat] = cat_counts.get(cat, 0) + 1
                    break
        
        categories_used = sorted(cat_counts.keys())
        print(f"   Categories: {', '.join(categories_used)}")

def _draw_mini_cpt(ax, required_capabilities, title, description):
    """Helper function to draw a miniature CPT for maturity comparison"""
    ax.set_xlim(0, 7)
    ax.set_ylim(0, 9)
    ax.set_aspect('equal')
    
    # Add title and description
    ax.text(3.5, 8.5, title, ha='center', va='center', 
           fontsize=12, fontweight='bold')
    ax.text(3.5, 8, description, ha='center', va='center', 
           fontsize=9, style='italic', color='#666666')
    
    # Remove axes
    ax.set_xticks([])
    ax.set_yticks([])
    for spine in ax.spines.values():
        spine.set_visible(False)
    
    # Draw capabilities
    for cap in capabilities:
        is_required = cap['id'] in required_capabilities
        colors = category_colors[cap['category']]
        alpha = 1.0 if is_required else 0.15
        
        x = cap['col'] - 0.4
        y = 9 - cap['row'] - 0.4
        
        rect = Rectangle((x, y), 0.8, 0.8, 
                        facecolor=colors['fill'], 
                        edgecolor=colors['border'],
                        linewidth=1.5 if is_required else 0.5,
                        alpha=alpha)
        ax.add_patch(rect)
        
        if is_required:
            ax.text(x + 0.4, y + 0.4, cap['id'], 
                   ha='center', va='center', 
                   fontsize=6, fontweight='bold',
                   color=colors['text'])

print("âœ… Maturity comparison function defined!")
````

## Step 9: PowerPoint Export

For business presentations, you often need PowerPoint format. The following function exports CPT visualizations directly to PowerPoint slides with proper formatting.

### Requirements:
```bash
pip install python-pptx
```

### Features:
- **Native PowerPoint format**: Editable shapes and text
- **Professional layout**: Proper sizing and positioning
- **Color consistency**: Matches visualization colors
- **Text formatting**: Proper fonts and sizing
- **Scalable**: Vector-based shapes for any resolution

````python
# PowerPoint Export Function

def export_to_powerpoint(config_name, save_path):
    """
    Export CPT to PowerPoint format using python-pptx
    
    Parameters:
    - config_name: predefined configuration to export
    - save_path: path for PowerPoint file (.pptx)
    """
    try:
        from pptx import Presentation
        from pptx.util import Inches, Pt
        from pptx.dml.color import RGBColor
        from pptx.enum.text import PP_ALIGN
        from pptx.enum.shapes import MSO_SHAPE
    except ImportError:
        print("âŒ PowerPoint export requires python-pptx")
        print("Install with: pip install python-pptx")
        return False
    
    if config_name not in configurations:
        print(f"âŒ Configuration '{config_name}' not found")
        return False
    
    config = configurations[config_name]
    
    # Create presentation
    prs = Presentation()
    
    # Create slide with blank layout
    slide_layout = prs.slide_layouts[6]  # Blank layout
    slide = prs.slides.add_slide(slide_layout)
    
    # Add title
    title_left = Inches(1)
    title_top = Inches(0.3)
    title_width = Inches(8)
    title_height = Inches(0.8)
    
    title_box = slide.shapes.add_textbox(title_left, title_top, title_width, title_height)
    title_frame = title_box.text_frame
    title_p = title_frame.paragraphs[0]
    title_p.text = config['title']
    title_p.font.size = Pt(24)
    title_p.font.bold = True
    title_p.alignment = PP_ALIGN.CENTER
    
    # Add description
    desc_box = slide.shapes.add_textbox(title_left, Inches(1), title_width, Inches(0.5))
    desc_frame = desc_box.text_frame
    desc_p = desc_frame.paragraphs[0]
    desc_p.text = config['description']
    desc_p.font.size = Pt(14)
    desc_p.font.color.rgb = RGBColor(102, 102, 102)
    desc_p.alignment = PP_ALIGN.CENTER
    
    # Add capabilities
    for cap in capabilities:
        is_required = cap['id'] in config['required']
        colors = category_colors[cap['category']]
        
        # Calculate position (scale and offset for PowerPoint)
        left = Inches(0.5 + (cap['col'] - 1) * 1.4)
        top = Inches(1.8 + (cap['row'] - 1) * 0.75)
        width = Inches(1.3)
        height = Inches(0.7)
        
        # Add rectangle shape
        shape = slide.shapes.add_shape(
            MSO_SHAPE.RECTANGLE, left, top, width, height
        )
        
        # Set fill color
        fill = shape.fill
        fill.solid()
        rgb_fill = _hex_to_rgb(colors['fill'])
        fill.fore_color.rgb = RGBColor(*rgb_fill)
        
        # Set transparency for non-required capabilities
        if not is_required:
            fill.transparency = 0.7
        
        # Set border
        line = shape.line
        rgb_border = _hex_to_rgb(colors['border'])
        line.color.rgb = RGBColor(*rgb_border)
        line.width = Pt(2 if is_required else 1)
        
        # Add text
        text_frame = shape.text_frame
        text_frame.margin_left = Inches(0.05)
        text_frame.margin_right = Inches(0.05)
        text_frame.margin_top = Inches(0.05)
        text_frame.margin_bottom = Inches(0.05)
        
        # Clear default paragraph
        text_frame.clear()
        
        # Add capability ID (bold)
        p1 = text_frame.paragraphs[0]
        p1.text = cap['id']
        p1.font.size = Pt(10)
        p1.font.bold = True
        rgb_text = _hex_to_rgb(colors['text'])
        p1.font.color.rgb = RGBColor(*rgb_text)
        p1.alignment = PP_ALIGN.CENTER
        
        # Add capability name
        p2 = text_frame.add_paragraph()
        p2.text = cap['name'].replace('\n', ' ')
        p2.font.size = Pt(8)
        p2.font.bold = is_required
        p2.font.color.rgb = RGBColor(*rgb_text)
        p2.alignment = PP_ALIGN.CENTER
    
    # Add legend
    legend_top = Inches(8.5)
    legend_categories = [
        ('PK', 'Perception & Knowledge'),
        ('CG', 'Cognition & Reasoning'), 
        ('LA', 'Learning & Adaptation'),
        ('AE', 'Action & Execution'),
        ('IC', 'Interaction & Collaboration'),
        ('GS', 'Governance & Safety')
    ]
    
    for idx, (cat, name) in enumerate(legend_categories):
        # Calculate position for 3x2 legend grid
        col = idx % 3
        row = idx // 3
        left = Inches(1 + col * 3)
        top = legend_top + row * Inches(0.4)
        
        # Add legend color box
        color_box = slide.shapes.add_shape(
            MSO_SHAPE.RECTANGLE, left, top, Inches(0.3), Inches(0.3)
        )
        colors = category_colors[cat]
        color_box.fill.solid()
        color_box.fill.fore_color.rgb = RGBColor(*_hex_to_rgb(colors['fill']))
        color_box.line.color.rgb = RGBColor(*_hex_to_rgb(colors['border']))
        
        # Add legend text
        text_box = slide.shapes.add_textbox(
            left + Inches(0.4), top, Inches(2.5), Inches(0.3)
        )
        text_frame = text_box.text_frame
        p = text_frame.paragraphs[0]
        p.text = f"{cat}: {name}"
        p.font.size = Pt(10)
        p.font.color.rgb = RGBColor(0, 0, 0)
    
    # Save presentation
    prs.save(save_path)
    print(f"âœ… PowerPoint exported to: {save_path}")
    return True

def _hex_to_rgb(hex_color):
    """Convert hex color to RGB tuple"""
    hex_color = hex_color.lstrip('#')
    return tuple(int(hex_color[i:i+2], 16) for i in (0, 2, 4))

print("âœ… PowerPoint export function defined!")
````

````python
# Export Configurations to PowerPoint

print("ðŸ“‹ Exporting configurations to PowerPoint format...")

# Export all predefined configurations
exported_files = []
for config_name in configurations.keys():
    filename = f"{config_name}_cpt.pptx"
    if export_to_powerpoint(config_name, filename):
        exported_files.append(filename)

if exported_files:
    print(f"\nâœ… Successfully exported {len(exported_files)} PowerPoint files:")
    for filename in exported_files:
        print(f"   ðŸ“„ {filename}")
else:
    print("\nâš ï¸  PowerPoint export failed. Install python-pptx to enable this feature.")
````

## Step 10: Interactive Capability Selection

For dynamic exploration, we can create an interactive widget that allows real-time capability selection and visualization.

### Features:
- **Checkbox interface**: Select/deselect capabilities by category
- **Live preview**: Instant visualization updates
- **Custom titles**: Editable title and description
- **Export options**: Save custom configurations

### Requirements:
```bash
pip install ipywidgets
jupyter nbextension enable --py widgetsnbextension
```

This is optional but provides a user-friendly interface for non-technical users.

````python
# Interactive Capability Selection Widget

def create_interactive_widget():
    """
    Create an interactive widget for capability selection
    Requires: pip install ipywidgets
    """
    try:
        import ipywidgets as widgets
        from IPython.display import display, clear_output
        
        # Create capability checkboxes organized by category
        capability_widgets = {}
        
        # Organize capabilities by category
        categories_organized = {}
        for cap in capabilities:
            cat = cap['category']
            if cat not in categories_organized:
                categories_organized[cat] = []
            categories_organized[cat].append(cap)
        
        # Create widgets for each capability
        category_widgets = {}
        for cat in ['PK', 'CG', 'LA', 'AE', 'IC', 'GS']:
            category_widgets[cat] = []
            for cap in categories_organized[cat]:
                widget = widgets.Checkbox(
                    value=False,
                    description=f"{cap['id']} - {cap['name'].replace(chr(10), ' ')}",
                    style={'description_width': '300px'},
                    layout={'width': '400px'}
                )
                capability_widgets[cap['id']] = widget
                category_widgets[cat].append(widget)
        
        # Create control widgets
        title_widget = widgets.Text(
            value='Custom CPT Configuration',
            description='Title:',
            style={'description_width': '100px'},
            layout={'width': '500px'}
        )
        
        desc_widget = widgets.Textarea(
            value='Custom capability selection for specific use case',
            description='Description:',
            style={'description_width': '100px'},
            layout={'width': '500px', 'height': '80px'}
        )
        
        # Preset configuration dropdown
        preset_widget = widgets.Dropdown(
            options=[('Select preset...', '')] + [(configurations[k]['title'], k) for k in configurations.keys()],
            description='Load Preset:',
            style={'description_width': '100px'},
            layout={'width': '400px'}
        )
        
        # Action buttons
        generate_button = widgets.Button(
            description='ðŸŽ¨ Generate CPT',
            button_style='success',
            layout={'width': '150px'}
        )
        
        clear_button = widgets.Button(
            description='ðŸ—‘ï¸ Clear All',
            button_style='warning',
            layout={'width': '150px'}
        )
        
        save_button = widgets.Button(
            description='ðŸ’¾ Save PNG',
            button_style='info',
            layout={'width': '150px'}
        )
        
        # Output area
        output = widgets.Output()
        
        # Event handlers
        def on_preset_change(change):
            if change['new'] and change['new'] in configurations:
                config = configurations[change['new']]
                title_widget.value = config['title']
                desc_widget.value = config['description']
                
                # Clear all checkboxes first
                for widget in capability_widgets.values():
                    widget.value = False
                
                # Check required capabilities
                for cap_id in config['required']:
                    if cap_id in capability_widgets:
                        capability_widgets[cap_id].value = True
        
        def on_generate_click(b):
            with output:
                clear_output(wait=True)
                selected = [cap_id for cap_id, widget in capability_widgets.items() 
                           if widget.value]
                
                if selected:
                    print(f"ðŸŽ¨ Generating CPT with {len(selected)} capabilities...")
                    create_cpt_visualization(
                        required_capabilities=selected,
                        title=title_widget.value,
                        description=desc_widget.value,
                        figsize=(14, 10)
                    )
                else:
                    print("âŒ Please select at least one capability!")
        
        def on_clear_click(b):
            for widget in capability_widgets.values():
                widget.value = False
            title_widget.value = 'Custom CPT Configuration'
            desc_widget.value = 'Custom capability selection for specific use case'
            preset_widget.value = ''
            with output:
                clear_output()
        
        def on_save_click(b):
            with output:
                selected = [cap_id for cap_id, widget in capability_widgets.items() 
                           if widget.value]
                
                if selected:
                    filename = f"custom_cpt_{len(selected)}_capabilities.png"
                    create_cpt_visualization(
                        required_capabilities=selected,
                        title=title_widget.value,
                        description=desc_widget.value,
                        figsize=(16, 12),
                        save_path=filename
                    )
                    print(f"ðŸ’¾ Saved as: {filename}")
                else:
                    print("âŒ Please select at least one capability!")
        
        # Connect event handlers
        preset_widget.observe(on_preset_change, names='value')
        generate_button.on_click(on_generate_click)
        clear_button.on_click(on_clear_click)
        save_button.on_click(on_save_click)
        
        # Create category tabs
        category_names = {
            'PK': 'Perception & Knowledge',
            'CG': 'Cognition & Reasoning',
            'LA': 'Learning & Adaptation',
            'AE': 'Action & Execution',
            'IC': 'Interaction & Collaboration',
            'GS': 'Governance & Safety'
        }
        
        tab_children = []
        tab_titles = []
        
        for cat in ['PK', 'CG', 'LA', 'AE', 'IC', 'GS']:
            tab_children.append(widgets.VBox(category_widgets[cat]))
            tab_titles.append(f"{cat}: {category_names[cat]}")
        
        tabs = widgets.Tab(children=tab_children)
        for i, title in enumerate(tab_titles):
            tabs.set_title(i, title)
        
        # Create main interface
        controls = widgets.VBox([
            widgets.HTML("<h3>ðŸŽ›ï¸ CPT Configuration Controls</h3>"),
            title_widget,
            desc_widget,
            preset_widget,
            widgets.HBox([generate_button, clear_button, save_button]),
            widgets.HTML("<br>")
        ])
        
        capabilities_section = widgets.VBox([
            widgets.HTML("<h3>ðŸ“‹ Select Capabilities by Category</h3>"),
            tabs
        ])
        
        main_interface = widgets.VBox([
            controls,
            capabilities_section,
            widgets.HTML("<h3>ðŸ“Š Generated Visualization</h3>"),
            output
        ])
        
        display(main_interface)
        
        print("âœ… Interactive widget created!")
        print("ðŸ’¡ Use the tabs above to select capabilities by category")
        print("ðŸŽ¯ Try loading a preset configuration to see examples")
        
    except ImportError:
        print("âŒ Interactive widgets require ipywidgets")
        print("Install with: pip install ipywidgets")
        print("Enable with: jupyter nbextension enable --py widgetsnbextension")
        return False
    
    return True

print("âœ… Interactive widget function defined!")
````
